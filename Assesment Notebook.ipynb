```python
df = spark.read.option("multiline", "true").json("/Volumes/assessment/default/assessment/Dataset.json")

display(df)
df.printSchema()
```

```python
from pyspark.sql.functions import explode

df_exploded = df.select(explode("entry").alias("entry"))

df_patient = df_exploded.select("entry.resource.*")

display(df_patient)
```

```python
from pyspark.sql.functions import current_timestamp, lit

df_bronze = df_patient.withColumn("extraction_timestamp", current_timestamp()) \
                      .withColumn("api_source", lit("FHIR_API"))

df_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_patient")
```

```python
display(spark.table("bronze_patient"))
```

```python
from pyspark.sql.functions import current_date

df_old = spark.table("dim_patient")

# Join old and new
df_join = df_old.alias("old").join(
    df_new.alias("new"),
    "id"
)

# Find changed records
df_changed = df_join.filter("old.gender != new.gender AND old.is_current = true")

# Expire old records
df_expired = df_old.join(
    df_changed.select("id"),
    "id"
).withColumn("end_date", current_date()) \
 .withColumn("is_current", lit(False))

# New records
df_insert = df_new.withColumn("start_date", current_date()) \
                  .withColumn("end_date", lit(None).cast("date")) \
                  .withColumn("is_current", lit(True))

# Final table
df_final = df_old.unionByName(df_insert)

df_final.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("dim_patient")
```

```python
%sql
CREATE OR REPLACE TABLE gold_patient_summary_full AS
SELECT 
    p.id AS patient_id,
    p.gender,
    COUNT(DISTINCT e.id) AS total_encounters,
    COUNT(DISTINCT o.id) AS total_observations,
    COUNT(DISTINCT c.id) AS total_conditions
FROM silver_patient p
LEFT JOIN silver_encounter e ON p.id = e.patient_id
LEFT JOIN silver_observation o ON p.id = o.patient_id
LEFT JOIN silver_condition c ON p.id = c.patient_id
GROUP BY p.id, p.gender;
```
